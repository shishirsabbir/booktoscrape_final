```python
import json
import csv
```

```python
data_list = []
with open('book_url_data.csv', mode='r', encoding='utf-8') as file:
    for line in csv.DictReader(file, fieldnames=('id', 'name', 'price', 'url')):
        data_list.append(line)
        print(line)
```

```python
with open('book_url_json_data.json', mode='w') as file:
    get_json = json.dumps(data_list)
    file.write(get_json)
    print('file written')
```
